```python
df.drop(df.index[(df['target']=='Irrelevant')],axis=0,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
df.drop(['id','location'],axis=1,inplace=True)
```

```python
df
```

```python
df['target']=df['target'].map({'Positive':1,'Negative':-1,'Neutral':0})
```

```python
df
```

```python
tweets=df.text
tweets
```

```python
tweets=tweets.str.replace('[^a-zA-Z0-9]+',' ')
tweets
```

```python
from nltk.stem import SnowballStemmer
sw=SnowballStemmer('english')
from nltk import TweetTokenizer
tk=TweetTokenizer()
tweets=tweets.apply(lambda x:[sw.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x: ' '.join(x))
```

```python
tweets
```

```python

from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
#tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i  not in stop].apply(lambda x: ' '.join(x)))
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in stop]).apply(lambda x:' '.join(x))
```

```python
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
train_data=vec.fit_transform(tweets)
```

```python
print(train_data)
```

```python
y=df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,y,test_size=0.30,random_state=1)
```

```python
from sklearn import svm
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
svm_model=svm.SVC()
mb=MultinomialNB()
rc=RandomForestClassifier()
aa=AdaBoostClassifier()
list_model=[svm_model,mb,rc,aa]
```

```python
from sklearn.metrics import classification_report
for i in list_model:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("******************************************")
  print(classification_report(y_test,y_pred))
  print("******************************************")
```
